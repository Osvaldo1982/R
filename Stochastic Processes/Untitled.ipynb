```r
trials <- 10000 # simulation repeated 10,000 times
simlist <- numeric(trials)
```

```r
toppings <- c("pepper","pepperoni","pineapple",
"prawns","prosciutto")
```

```r
toppings
```

```r
for (i in 1:trials) 
    {
pineapple <- 0
pepperoni <- 0 # counts pepperonis before pineapple
while (pineapple == 0) 
    {

    # pick toppings until pineapple is selected
pick <- sample(toppings,1)
if (pick == "pepperoni") pepperoni <- pepperoni + 1
if (pick == "pineapple") pineapple <- 1 
    }
simlist[i] <- pepperoni 
    }
```

```r
mean(simlist)
```
