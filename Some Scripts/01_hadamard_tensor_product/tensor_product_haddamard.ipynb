## Defining a null matrix

```r
#
x <- c(1,1)

#
n <- length(x)

#
k <- 5

#
res <- matrix(0, ncol = k * n, nrow = k)

#
res
```

```r
for (i in 1:k){
  res[i,((i - 1)*n + 1):(i*n)] <- x
  }

res
```

## Kronecker product

In mathematics, the Kronecker product, sometimes denoted by âŠ—, is an operation on two matrices of arbitrary size resulting in a block matrix. It is a generalization of the outer product (which is denoted by the same symbol) from vectors to matrices, and gives the matrix of the tensor product linear map with respect to a standard choice of basis. The Kronecker product is to be distinguished from the usual matrix multiplication, which is an entirely different operation. The Kronecker product is also sometimes called matrix direct product.
